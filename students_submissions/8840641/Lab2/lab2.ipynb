# Making the PIE chart for %of land occupied by each continents on earth

```python
import matplotlib.pyplot as plt

# Continent names and their respective land percentages
continents = ["Asia", "Africa", "North America", "South America", "Europe", "Australia", "Antarctica"]
land_percentages = [30, 20, 16, 15, 10, 8, 1]  # Percentages of land for each continent

# Define colors for each continent
colors = ["#FF9999", "#66B2FF", "#99FF99", "#FFCC99", "#c2c2f0", "#ffb3e6", "#66b070"]

# Create a pie chart with custom colors
plt.figure(figsize=(8, 8))
plt.pie(land_percentages, labels=continents,colors=colors,autopct='%1.1f%%')

# Set the title
plt.title("Percentage of Land Area by Continent")

# Display the pie chart
plt.show()
```

# Bar graph for number of people speaking different languages in India 

```python
import matplotlib.pyplot as plt
import numpy as np

# Languages and the number of people who speak them in crores
languages = ["Hindi", "Gujarati", "Marathi", "Telugu", "Tamil", "English","Other"]
speakers_in_million = [20, 5, 10, 2, 5, 1, 25]

# Create a bar graph
plt.figure(figsize=(12, 6))
plt.bar(languages, speakers_in_million, color='teal')

# Adding labels and title
plt.xlabel("Languages",fontweight='bold',fontstyle='italic')
plt.ylabel("Population (in Million)",fontweight='bold',fontstyle='italic')
plt.title("Number of People Speaking Different Languages in India",fontweight='bold',fontstyle='italic')

# Display the bar graph
plt.show()
```

# Highest and Lowest Temperature in °C in Kitchener, Ontario

```python
import calendar
import matplotlib.pyplot as plt
import numpy as np

month =(["January","February", "March", "April", "May", "June", "July", "August", "September", "October", "November", "December"]) #x- axis
# y-axis (Derived from https://www.currentresults.com/Weather/Canada/Ontario/Places/kitchener-waterloo-temperatures-by-month-average.php)
temp_low = np.array([-11, -11, -6, -1, 7,11,14,13,9,4,-2,-6]) 
temp_high = np.array([-2,-2,4,11,20,24,28,26,22,15,7,1])


#Creating plot chart
plt.figure(figsize=(12,6))
plt.plot(month, temp_low, label="Lowest Temperature", color='blue',marker='o')
plt.plot(month, temp_high, label="Highest  Temperature", color='red',marker='o')

#Adding Labels and Titles
plt.title("Average Temperature in Kitchener, Ontario  every month")
plt.xlabel("Month")
plt.ylabel("Temperature")

plt.legend() #Adding legend

plt.grid() # Add grid view

plt.show()
```

```python
import matplotlib.pyplot as plt
import matplotlib
matplotlib.rcParams['font.size'] = 8.0

# Fixing random state for reproducibility
np.random.seed(239688991)


# create random data
data1 = np.random.random([6, 50])

# set different colors for each set of positions
colors1 = [f'C{i}' for i in range(6)]

# set different line properties for each set of positions
# note that some overlap
lineoffsets1 = [-15, -3, 1, 1.5, 6, 10]
linelengths1 = [5, 2, 1, 1, 3, 1.5]

fig, axs = plt.subplots(2, 2)

# create a horizontal plot
axs[0, 0].eventplot(data1, colors=colors1, lineoffsets=lineoffsets1,
                    linelengths=linelengths1)

# create a vertical plot
axs[1, 0].eventplot(data1, colors=colors1, lineoffsets=lineoffsets1,
                    linelengths=linelengths1, orientation='vertical')

# create another set of random data.
# the gamma distribution is only used for aesthetic purposes
data2 = np.random.gamma(4, size=[60, 50])

# use individual values for the parameters this time
# these values will be used for all data sets (except lineoffsets2, which
# sets the increment between each data set in this usage)
colors2 = 'black'
lineoffsets2 = 1
linelengths2 = 1

# create a horizontal plot
axs[0, 1].eventplot(data2, colors=colors2, lineoffsets=lineoffsets2,
                    linelengths=linelengths2)


# create a vertical plot
axs[1, 1].eventplot(data2, colors=colors2, lineoffsets=lineoffsets2,
                    linelengths=linelengths2, orientation='vertical')

plt.show()
```
